```python
#Missing values
features_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>=1]
```

```python
for features in features_nan:
    print("The feature is {} and missing value in {}%".format(features,np.round(dataset[features].isnull().mean()*100,4)))
```

```python
#Replacing categorical feature--
categorical_feature=[feature for feature in dataset.columns if dataset[feature].dtypes=='O']
len(categorical_feature)
```

```python
# nan value replace with missing_value
data = dataset.copy()
def replace_categorical_feature(dataset,categorical_feature):
    data[categorical_feature] = data[categorical_feature].fillna('Missing_value')
    return data
```

```python
replace_categorical_feature(data,categorical_feature)
```

```python
# nan value replace with Mode
data = dataset.copy()
def replace_with_mode_feature(dataset,categorical_feature):
    data[categorical_feature] = data[categorical_feature].fillna(data[categorical_feature].mode())
    return data
```

```python
replace_with_mode_feature(data,categorical_feature)
```

```python
numerical_features_nan = [feature for feature in dataset.columns if dataset[feature].dtype!='O' and dataset[feature].isnull().sum()>=1]
```

```python
numerical_features_nan
```

```python
#replacing numerical_features_nan with median value

for feature in numerical_features_nan:
    #since there are outliers we r going to replace with median
    median_values = dataset [feature].median()
    #create a new feature to capture nan value
    dataset[feature+'nan'] = np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_values,inplace = True)
```

```python
dataset.head()
```

```python
#temporal variables (date time variables)

for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    dataset[feature +'age of house']= dataset['YrSold']-dataset[feature]
```
